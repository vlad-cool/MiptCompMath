```python
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

# Исследуемые функции:

```python
def f_1(x):
    return math.sin(x * x)

def f_1_derivative(x):
    return 2 * x * math.cos(x * x)

f_1.derivative = f_1_derivative
f_1.expression = "\\sin(x^2)"

def f_2(x):
    return math.cos(math.sin(x))

def f_2_derivative(x):
    return -math.sin(math.sin(x)) * math.cos(x)

f_2.derivative = f_2_derivative
f_2.expression = "\\cos(\\sin(x))"

def f_3(x):
    return math.exp(math.sin(math.cos(x)))

def f_3_derivative(x):
    return -f_3(x) * math.cos(math.cos(x)) * math.sin(x)

f_3.derivative = f_3_derivative
f_3.expression = "\\exp(\\sin(\\cos(x)))"

def f_4(x):
    return math.log(x + 3)

def f_4_derivative(x):
    return 1 / (x + 3)

f_4.derivative = f_4_derivative
f_4.expression = "\\ln(x + 3)"

def f_5(x):
    return math.pow(x + 3, 0.5)

def f_5_derivative(x):
    return 1 / (2 * f_5(x))

f_5.derivative = f_5_derivative
f_5.expression = "(x + 3)^{0.5}"

functions = [f_1, f_2, f_3, f_4, f_5]
```

# Формулы производных

```python

def derivative_1(f, x, h):
    return (f(x + h) - f(x)) / h

derivative_1.expression = "\\frac{f(x + h) - f(x)}{h}"

def derivative_2(f, x, h):
    return (f(x) - f(x - h)) / h

derivative_2.expression = "\\frac{f(x) - f(x - h)}{h}"

def derivative_3(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

derivative_3.expression = "\\frac{f(x + h) - f(x - h)}{2h}"

def derivative_4(f, x, h):
    return 4 * (f(x + h) - f(x - h)) / (6 * h) - (f(x + 2 * h) - f(x - 2 * h)) / (12 * h)

derivative_4.expression = "\\frac{4}{3} \\ \\frac{f(x + h) - f (x - h)}{2h} - \\frac{1}{3} \\ \\frac{f(x + 2h) - f(x - 2h)}{4h}"

def derivative_5(f, x, h):
    return 3 * (f(x + h) - f(x - h)) / (4 * h) - 3 * (f(x + 2 * h) - f(x - 2 * h)) / (20 * h) + (f(x + 3 * h) - f(x - 3 * h)) / (60 * h)

derivative_5.expression = "\\frac{3}{2} \\ \\frac{f(x + h) - f(x - h)}{2h} - \\frac{3}{5} \\ \\frac{f(x + 2h) - f(x - 2h)}{4h} + \\frac{1}{10} \\ \\frac{f(x + 3h) - f(x - 3h)}{6h}"

derivatives = [derivative_1, derivative_2, derivative_3, derivative_4, derivative_5]
```

# Значение $x_0$, для которого вычисляется производная

```python
x_0 = math.pi / 3
```

```python
for function in functions:
    for derivative in derivatives:
        x = []
        y = []

        for n in range(1, 22):
            h = 2 / (2 ** n)
            x.append(n)
            y.append(abs(derivative(function, x_0, h) - function.derivative(x_0)))
        plt.plot(x, y, label=f"$f'(x)={derivative.expression}$")
    plt.xscale("log")
    plt.yscale("log")
    plt.xlabel("$n$ - шаг численного дифференцирования")
    plt.ylabel("$\\Delta f'$ - погрешность производной")
    plt.title(f"График погрешностей для производной функции $f(x) = {function.expression}$")
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2))
    plt.show()
```
